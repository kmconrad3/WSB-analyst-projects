```python
df = df[["Term", "Campus ID", "Academic Plan Code", "Academic Plan"]]
df = pd.get_dummies(df, columns=["Academic Plan Code"]).drop("Academic Plan", axis=1)

#majs =df.columns.tolist()[2:]
#for col in majs:
#    df.rename(columns={col : (col.split("Code_",1)[1])}, inplace = True)
 
    
# figure out how to not hard code    
df.rename(columns={ "Academic Plan Code_BUS 004": "BUS: Accounting", \
                   "Academic Plan Code_BUS 009": "BUS: Actuarial Science", \
                   "Academic Plan Code_BUS 423": "BUS: Finance, Invest&Bank", \
                   "Academic Plan Code_BUS 569": "BUS: Information System", \
                   "Academic Plan Code_BUS 575": "BUS: International Bus", \
                   "Academic Plan Code_BUS 623": "BUS: Management & Hum Res", \
                   "Academic Plan Code_BUS 625": "BUS: Marketing", \
                   "Academic Plan Code_BUS 846": "BUS: Oper & Tech Mgm", \
                   "Academic Plan Code_BUS 859": "BUS: RealEst & UrbLandEcon", \
                   "Academic Plan Code_BUS 882": "BUS: Risk Mgmt & Insurance", \
                   "Academic Plan Code_BUS 986": "BUS: Supply Chain Mgmt", \
                   "Academic Plan Code_MAJ 229": "Computer Sciences", \
                   "Academic Plan Code_MAJ 252": "Data Science", \
                   "Academic Plan Code_MAJ 261": "Economics", \
                   "Academic Plan Code_MAJ 634": "Mathematics", \
                   "Academic Plan Code_MAJ 949": "Statistics"}, inplace = True)


df = df.groupby(["Campus ID", "Term"]).agg(func="max")
df = df.reset_index()

#Checking
#df[df["Campus ID"] == 9069101153]
```

# For relations table

```python
df2 = pd.read_csv("Student Academic Plans.csv", low_memory=False)
df3 = pd.read_csv("Student Academic Plans.csv", low_memory=False)
df2 = df2[["Term", "Campus ID", "Academic Plan"]]
df3 = df3[["Term", "Campus ID", "Academic Plan"]]

df2 = df2.replace("Computer Sciences BMAJ", "Computer Sciences")
df3 = df3.replace("Computer Sciences BMAJ", "Computer Sciences")

df2 = df2.replace("Economics BMAJ", "Economics")
df3 = df3.replace("Economics BMAJ", "Economics")

df2 = df2.replace("Mathematics BMAJ", "Mathematics")
df3 = df3.replace("Mathematics BMAJ", "Mathematics")

df2 = df2.replace("Statistics BMAJ", "Statistics")
df3 = df3.replace("Statistics BMAJ", "Statistics")

df3 = df3.rename( columns={ "Academic Plan": "Academic Plan 2"})
df2 = df2.merge(df3, how="outer")

#Checking
#pd.set_option('display.max_rows', 1000)
#df2[df2["Campus ID"] == 9084935114]
```
