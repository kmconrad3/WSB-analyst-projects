## Not split data

```python
label_dict = pd.read_csv("Map_dict_for_SPLIT_Q_names_CSV.csv")

surv_2017= pd.read_csv("All years surv data/BBA_Program_Survey_2017_Not_Split.csv")
surv_2018= pd.read_csv("All years surv data/BBA_Program_Survey_2018_Not_Split.csv")
surv_2019= pd.read_csv("All years surv data/BBA_Program_Survey_2019_Not_Split.csv")
surv_2020= pd.read_csv("All years surv data/BBA_Program_Survey_2020_Not_Split.csv")
surv= pd.read_csv("All years surv data/BBA_Program_Survey_Not_Split.csv")

all_survs = [surv_2017, surv_2018, surv_2019, surv_2020, surv]
surv_names = ["surv_2017", "surv_2018", "surv_2019", "surv_2020", "surv"]

droppings={}
for ro, name in zip(all_survs, surv_names):
    to_drop=[]
    for num_col in range(len(ro.columns)):
        if ro.columns[num_col] in label_dict[name].values:
            find = ro.columns[num_col]
            ind = label_dict[label_dict[name] == find].index[0]
            ro.rename(columns={find: label_dict.loc[ind, "New Name"]}, inplace=True)
        else:
            if ro.columns[num_col] not in to_drop:
                to_drop.append(ro.columns[num_col])
                
    #info about dropped questions
    print("{}: {}, {}".format(name, len(ro.columns), len(to_drop)) )
    droppings[name]=to_drop
    ro.drop(to_drop, axis=1, inplace=True)
```

```python
surv = pd.concat([surv, surv_2020, surv_2019, surv_2018, surv_2017])
surv.to_excel("all_surv_renamed.xlsx", "all_surv_renamed.xlsx")
#droppings
```

```python
#Seeing if we missed an columns by checking title strings similarity
from difflib import SequenceMatcher
def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

for x in surv.columns:
    for i in droppings:
        for ii in droppings[i]:
            #print(ii)
            if similar(ii, x)> .6:
                pass
                #print(i + ": " + ii + "&&" + x)
                #print(" ")
```
